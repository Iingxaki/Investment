```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of Trading Economics Economic Calendar
url = "https://tradingeconomics.com/calendar"

# Add headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Fetch the webpage with headers
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Extract economic event data
events = []
# Find all rows with class 'calendar-row'
for row in soup.find_all("tr", class_="calendar-row"):
    # Extract data from each column
    time = row.find("td", class_="calendar-date").text.strip()
    headline = row.find("td", class_="calendar-event").text.strip()
    impact = row.find("td", class_="calendar-impact").text.strip()
    actual = row.find("td", class_="calendar-actual").text.strip()
    previous = row.find("td", class_="calendar-previous").text.strip()
    consensus = row.find("td", class_="calendar-consensus").text.strip()
    forecast = row.find("td", class_="calendar-forecast").text.strip()
    category = row.find("td", class_="calendar-country").text.strip()


    events.append({
        "Time": time,
        "Headline": headline,
        "Impact": impact,
        "Actual": actual,
        "Previous": previous,
        "Consensus": consensus,
        "Forecast": forecast,
        "Category": category
    })

# Convert to DataFrame
df_events = pd.DataFrame(events)

# Display the first few rows
print(df_events.head())
```

```python
# prompt: use playwright to do headless browsing into tradeeconomics.com/calendar and scrape all its news headlines from all countries related to EUDJPY, USDJPY, AUDUSD. have the columns be date, time of day UTC-4, impact rating, currency pair it relates to, actual, previous, consensus, forecast.

import asyncio
from playwright.async_api import async_playwright
import pandas as pd

async def scrape_tradingeconomics_calendar():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        await page.goto("https://tradingeconomics.com/calendar", wait_until='networkidle')

        # Click "ALL COUNTRIES" to ensure all data is loaded (assuming this button exists and expands the list)
        try:
            await page.click('button:has-text("ALL COUNTRIES")')
            # Wait for the list to potentially update or expand after clicking
            await page.wait_for_timeout(2000) # Wait for 2 seconds, adjust if needed
        except:
            print("Could not find or click 'ALL COUNTRIES' button. Proceeding with visible data.")
            pass # Continue if the button is not found

        # Locate the table rows containing the calendar data
        rows = await page.locator('table.calendar-table tbody tr.calendar-row').all()

        events_data = []
        for row in rows:
            # Extract data from relevant columns
            # Use .text_content() to get all text within the element
            # Use .get_attribute('class') to check for impact rating
            date_time_utc = await row.locator('.calendar-date').text_content()
            country = await row.locator('.calendar-country').text_content()
            headline = await row.locator('.calendar-event').text_content()

            # Determine impact rating based on the class of the impact cell
            impact_element = row.locator('.calendar-impact')
            impact_classes = await impact_element.get_attribute('class')
            impact_rating = 'Unknown'
            if 'calendar-impact-high' in impact_classes:
                impact_rating = 'High'
            elif 'calendar-impact-medium' in impact_classes:
                impact_rating = 'Medium'
            elif 'calendar-impact-low' in impact_classes:
                impact_rating = 'Low'

            actual = await row.locator('.calendar-actual').text_content()
            previous = await row.locator('.calendar-previous').text_content()
            consensus = await row.locator('.calendar-consensus').text_content()
            forecast = await row.locator('.calendar-forecast').text_content()

            # Simple check for relevant currency pairs in the headline
            # This is a basic approach and might need refinement based on actual data
            currency_pairs = []
            if 'EUDJPY' in headline or 'Euro' in headline or 'Japan' in headline:
                currency_pairs.append('EUDJPY')
            if 'USDJPY' in headline or 'United States' in headline or 'Japan' in headline:
                currency_pairs.append('USDJPY')
            if 'AUDUSD' in headline or 'Australia' in headline or 'United States' in headline:
                currency_pairs.append('AUDUSD')

            # Add to data if any relevant currency pair is found
            if currency_pairs:
                events_data.append({
                    'Date_Time_UTC': date_time_utc.strip(),
                    'Country': country.strip(),
                    'Headline': headline.strip(),
                    'Impact Rating': impact_rating,
                    'Currency Pair': ', '.join(currency_pairs), # Join pairs if multiple match
                    'Actual': actual.strip(),
                    'Previous': previous.strip(),
                    'Consensus': consensus.strip(),
                    'Forecast': forecast.strip()
                })

        await browser.close()
        return pd.DataFrame(events_data)

# Run the async function
df_calendar_data = await scrape_tradingeconomics_calendar()

# Display the DataFrame
print(df_calendar_data.shape)
display(df_calendar_data.head())
```
